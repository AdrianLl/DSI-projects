```python
import pandas as pd
import cPickle
import time
import datetime
from geopy.geocoders import Nominatim
from geopy.distance import great_circle
```

```python
df = cPickle.load(open('assets/big_df.p', 'rb')) 
df.reset_index(drop = True, inplace = True)
```

```python
start = datetime.datetime.now()
print 'start time: %s' %start
location = []
#    location_info = pd.DataFrame()
attention = []
geolocator = Nominatim()
count = -1
for i in range(len(df) - len(location)):
    coordinates = str(df['latitude'][i]) + ', ' + str(df['longitude'][i])
    try:
        lil_location = geolocator.reverse(coordinates, timeout = 30)
        time.sleep(1.25)
        if not lil_location:
            print 'oops',
            location.append(None)
#            lat.append('fill me in!')
#            lon.append('fill me in!')
#            oops_index = df_NYC[df_NYC['address'] == street].index[0]
#            oops_location = pd.DataFrame([0]*len(location_info.columns)).T
#            location_info = pd.concat([location_info, oops_location], axis = 0)
            attention.append(df[df['latitude'] == df['latitude'][i]].index[0])
        else:
            location.append(lil_location.raw)
#             location.append(lil_location.address)
#             lat.append(float(lil_location.latitude))
#             lon.append(float(lil_location.longitude))
#             location_info = pd.concat([location_info, pd.DataFrame(lil_location.raw)], axis = 0)
        count += 1
        if count%500 == 0:
            print round(float(count)/len(df), 3),

    except:
        location.append(None)
        count += 1
        print '.',
        if count%500 == 0:
            print round(float(count)/len(df), 3),              

if len(df) == len(location):
    df['location'] = location
#     df['lat'] = lat
#     df['lon'] = lon

else:
    print '\n'
    print 'These did not geocode properly, please impute:\n'
    print df[['name', 'currency', 'blurb']].ix[attention]

end = datetime.datetime.now()
print 'done!  time elapsed: %s' %(end - start)
```

```python
df = df[(df['longitude'] > -74.4) & \
            (df['longitude'] < -73.6) & (df['latitude'] > 40.4) & \
            (df['latitude'] < 41)]
df.reset_index(drop = True, inplace = True)
```
