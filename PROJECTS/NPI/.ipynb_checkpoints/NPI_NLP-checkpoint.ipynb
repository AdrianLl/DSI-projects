```python
train_text = df['articles'][0]

sample_text = df['articles'][1]

custom_sent_tokenizer = PunktSentenceTokenizer(train_text.decode('utf-8'))

tokenized = custom_sent_tokenizer.tokenize(sample_text.decode('utf-8'))


def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            
            chunkGram = r'''Chunk: {<RB.?>*<VB.?>*<NN>?}'''
            
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            
            
            chunked.draw()
            
    except Exception as e:
        print str(e)
    
process_content()
```

```python
def senttoke_articles(df):
    senttoke_article = []
    for i in range(len(df)):
        df['articles'][i] = df['articles'][i].replace('.”', '”.')
        #df['articles'][i] = df['articles'][i].replace('[', ' ')  #[9]  9.
        #df['articles'][i] = df['articles'][i].replace(']', ' ')
        df['articles'][i] = df['articles'][i].replace('_', '')
        df['articles'][i] = df['articles'][i].replace('…', '')

        art = sent_tokenize(df['articles'][i].decode('utf-8'))
        
        senttoke_article.append(art)
        
    return senttoke_article
```

```python
df['senttokes'] = senttoke_articles(df)
```
