```python
gross = r_soup.find(text=re.compile('Domestic Total Gross'))
```

```python
distributor = r_soup.find(text=re.compile('Distributor'))
```

```python
rating = r_soup.find(text=re.compile('MPAA Rating'))
```

```python
rating.findNextSibling().text
```

```python
for link in r_soup.findAll('b'): # get link and text from inspect in browser
    print link.text
    #print link['href']
    print '\n'
```

```python
r_soup.findAll('b')[1].text
```

```python
my_dict = {
    'title' : r_soup.findAll('b')[1].text,
    'domestic gross' : gross.findNextSibling().text,
    'distributor' : distributor.findNextSibling().text,
    'MPAA rating' : rating.findNextSibling().text}
```

```python
df = pd.DataFrame(my_dict.items(), columns = ['category', 'info'])
```
