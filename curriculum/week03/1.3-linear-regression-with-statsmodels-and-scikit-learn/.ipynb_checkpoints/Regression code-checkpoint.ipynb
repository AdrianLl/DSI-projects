```python
import numpy as np
import pandas as pd
from pandas import Categorical as cat
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
import seaborn as sns
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
#(1, 3.0) (2, 5.1) (3, 7.1) (4, 9.4) (5, 11.5) (6, 13.5) (7, 15.6) (8, 17.7) (9, 19.8)
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [3, 5.1, 7.1, 9.4, 11.5, 13.5, 15.6, 17.7, 19.8]
```

```python
def regress(x, y):
    x_bar = float(sum(x))/len(x)
    y_bar = float(sum(y))/len(y)
    xy_bar = sum([x[i]*y[i] for i in range(len(x))])/len(x)
    x_sq_bar = sum([float(x[i])**2 for i in range(len(x))])/len(x)
    beta_hat = (xy_bar - (x_bar * y_bar)) / (x_sq_bar - x_bar**2)
    alpha = y_bar - beta_hat * x_bar

    print "x_bar: %s" %x_bar
    print "y_bar: %s" %y_bar
    print "xy_bar: %s" %xy_bar
    print "x^2 bar: %s" %x_sq_bar
    print "Beta hat: %s" %beta_hat
    print "Alpha intercept: %s" %alpha

    plt.figure(figsize=(12,8))
    plt.style.use('fivethirtyeight')

    df_scatter = plt.scatter(x, y, alpha=0.5, color = "darkblue", s = 20)

    plt.title('Regression of x, y', fontsize = 24)

    plt.ylabel('y')

    plt.xlabel('x')
    
    ablineValues = []
    for i in x:
        ablineValues.append(beta_hat*i+alpha)
    
    plt.plot(x, ablineValues, linewidth = .4, color = "darkgreen")
    plt.show()
    
    return
    
```

```python
#(-1, 0) (1, 3) (2, 4) (3, 7) (4, 10) 
x1 = [-1, 1, 2, 3, 4]
y1 = [0, 3, 4, 7, 10]
```

```python
regress(x1, y1)
```
